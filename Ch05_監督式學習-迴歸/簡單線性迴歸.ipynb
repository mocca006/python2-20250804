```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)

# prices of the pizzas.
y = [7, 9, 13, 17.5, 18]

plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(X, y, 'k.')
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y = [7, 9, 13, 17.5, 18]

model = LinearRegression()
model.fit(X, y)

test_pizza = np.array([[18]])
predicted_price = model.predict(test_pizza)[0]
print('12吋 pizza 應售: $%.2f 元' % predicted_price)
```

```python
print(model.coef_, model.intercept_)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y = [7, 9, 13, 17.5, 18]

plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(X, y, 'ko')
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.plot([0, 25], [1.97, 26.37], color='b', linestyle='-',
         linewidth=2)
plt.show()
```

```python
plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('size')
plt.ylabel('Price in dollars')
plt.plot(X, y, 'ro')
plt.axis([0, 25, 0, 25])
plt.grid(True)
plt.plot([0, 25], [0, 28], color='k', linestyle='-', 
         linewidth=2)
plt.plot([0, 25], [1, 29], color='b', linestyle='--', 
         linewidth=2)
plt.plot([-1, 25], [0.5, 27], color='c', linestyle='-.', 
         linewidth=2)
plt.show()
```
