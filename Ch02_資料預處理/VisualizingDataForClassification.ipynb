## Visualize class separation by numeric features 特徵視覺化

```python
def plot_box(credit, cols, col_x = 'bad_credit'): # 繪製箱型圖函數,creadit是資料表,cols是欄位名稱list, col_x是x軸欄位名稱
    for col in cols:
        sns.set_style("whitegrid") 
        # 在 Python 的 Seaborn 資料視覺化中，sns.set_style("whitegrid") 是用來設定繪圖的整體風格，指定背景為白色並加上網格線，這有助於增強圖表的可讀性。
        sns.boxplot(x=col_x, y=col, data=credit) # 繪製箱型圖
        plt.xlabel(col_x) # Set text for the x axis
        plt.ylabel(col) # Set text for y axis
        plt.show() # 顯示圖形

num_cols = ['loan_duration_mo', 'loan_amount', 'payment_pcnt_income',
            'age_yrs', 'number_loans'] # 繪製數值型欄位的箱型圖,欄位名稱list,會有5張圖表
plot_box(credit, num_cols) # 繪製數值型欄位的箱型圖

# 圖表一
# 圈圈是異常值
# 橫線是max與min
# 藍色框框是資料集中範圍,75%到25%
# 中間橫線是中位數50%

# 圖表三,四,五有些異常值
# 圖表一,二,三,四,五的中位數都偏低,資料分布有些偏斜
# 資料分布有些偏斜,有些欄位有異常值,可以考慮做標準化或正規化
# 可以考慮刪除異常值,或用中位數取代
# 可以考慮用對數轉換,或Box-Cox轉換,或Yeo-Johnson轉換
# 可以考慮用分箱,或分群,或決策樹,或隨機森林,或XGBoost,或LightGBM,或CatBoost
```

```python
def plot_violin(credit, cols, col_x = 'bad_credit'): 
    # 繪製小提琴圖函數,creadit是資料表,cols是欄位名稱list, col_x是x軸欄位名稱
    for col in cols:
        sns.set_style("whitegrid")
        #sns.violinplot(x=col_x, y=col, data=credit) # 繪製小提琴圖
        sns.violinplot(x=col_x, y=col, data=credit,hue='foreign_worker',split=True) 
         # 分割小提琴圖,hue是分組欄位
         # hue='foreign_worker',split=True 參數,將資料依照foreign_worker欄位分組,並且在同一個小提琴圖中顯示
         # split=True 參數,將資料分割成兩半,分別顯示在小提琴圖的左半部和右半部
         
        plt.xlabel(col_x) # Set text for the x axis
        plt.ylabel(col) # Set text for y axis
        plt.show()

plot_violin(credit, num_cols)
```

The interpretation of these plots is largely the same as the box plots. However, there is one detail worth noting. The differences between loan_duration_mo and loan_amount for good and bad credit customers are only for the more extreme values. It may be that these features are less useful and the box plot indicates.
