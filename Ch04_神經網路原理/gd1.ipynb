```python
# %load 'gd1.py'
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 目標函數:y=x^2
# 曲線
def func(x): 
    return np.square(x)

# 目標函數一階導數:dy/dx=2*x
# 切線
def dfunc(x): 
    return 2 * x
```

```python

def GD(x_start, df, epochs, lr):    
    """  梯度下降法。給定起始點與目標函數的一階導函數，求在epochs次反覆運算中x的更新值
        :param x_start: x的起始點    
        :param df: 目標函數的一階導函數    
        :param epochs: 反覆運算週期    
        :param lr: 學習率    
        :return: x在每次反覆運算後的位置（包括起始點），長度為epochs+1    
     """    
    xs = np.zeros(epochs+1) # 用來存放每次反覆運算後的x值    
      
    x = x_start # 初始點     
    xs[0] = x # 起始點    
     # epochs次反覆運算   
    for i in range(epochs):         
        dx = df(x) # 取得x在目前位置的導數值       
        # v表示x要改變的幅度        
        v = - dx * lr # 學習率lr控制每次改變的幅度,負號表示往導數的反方向移動       
        x += v #  更新x的位置        
         # 紀錄x的位置       
        xs[i+1] = x    
    return xs # 回傳x在每次反覆運算後的位置（包括起始點），長度為epochs+1
```

```python
# Main
# 起始權重
x_start = -5    
# 執行週期數
epochs = 20 
# 學習率   
lr = 0.1 # 學習率太大會震盪，太小會收斂很慢
# x的更新位置
# x = np.array([x_start])
# 梯度下降法 
# *** Function 可以直接當參數傳遞 ***
x = GD(x_start, dfunc, epochs, lr=lr) 
print (x)

```

```python

color = 'r'    
#plt.plot(line_x, line_y, c='b')    
from numpy import arange
t = arange(-6.0, 6.0, 0.01) # x軸座標
plt.plot(t, func(t), c='b') # 目標函數曲線
plt.plot(x, func(x), c=color, label='lr={}'.format(lr))  # 每次更新的位置  
plt.scatter(x, func(x), c=color, )  # 每次更新的位置  
plt.legend() # 顯示圖例

plt.show() # 
```
