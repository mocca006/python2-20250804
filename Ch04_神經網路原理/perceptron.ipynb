## 感知器
感知器的主要功能是處理二元分類問題，例如將數據分為「A 類」和「B 類」。雖然感知器的設計相對簡單，僅能處理線性可分的數據，但它為神經網路的發展奠定了理論基礎。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Feature values
X = np.array([[1,3,3],
            [1,4,31],
            [1,1,1],
            [1,0,2]])
# Target values
y = np.array([[1],
            [1],
            [-1],
            [-1]])
```

```python
# w是權重
w = (np.random.random([3,1])-0.5)*2 # 3*1個亂數
w
```

```python
a = np.dot(X, w) # 4*3 乘 3*1 = 4*1 
# a=X與w 矩陣內積
a
```

```python
print(np.sign(np.dot(X, w))) # 4*3 乘 3*1 = 4*1
# X與w 矩陣內積， np.sign() 判斷值的正負號：正為1.，負為-1.，0為0
# np.sign() 判斷值的正負號：正為1.，負為-1.，0為0
# Sigmoid函數 # 1/(1+np.exp(-a))
# lr = 0.1 # 學習率
# print(y)
```

## 梯度下降


```python
lr =0.11 # 學習率
out = 0

def update():
    global X, y, w,lr
    # 判斷值的正負號：正為1.，負為-1.，0為0
    out = np.sign(np.dot(X, w)) # 4*3 乘 3*1 = 4*1
    # 權重更新
    w_c = lr * (X.T.dot(y-out))/int(X.shape[0]) # 3*4 乘 4*1 = 3*1, X.shape[0] 代表X的列數
    # X.T 代表X的轉置矩陣
    # int(X.shape[0]) 代表X的列數
    # X.T.dot(y-out) 代表X的轉置矩陣與(y-out)的矩陣內積
    # (y-out) 代表目標值與預測值的差
    # lr * (X.T.dot(y-out))/int(X.shape[0]) 代表學習率乘以X的轉置矩陣與(y-out)的矩陣內積再除以X的列數
    w = w + w_c # 權重更新
```

```python
for i in range(20): # 迭代20次
    update()
    x1=[3,4]
    y1=[3,3]
    x2=[1,0]
    y2=[1,2]

    k = -w[1]/w[2]
    d = -w[0]/w[2]
    print('k=',k)
    print('d=',d)

    xdata = (0,5)
    plt.figure()
    plt.plot(xdata, xdata*k+d,'r')
    plt.scatter(x1,y1,c='b')
    plt.scatter(x2,y2,c='y')
    plt.show()
```
