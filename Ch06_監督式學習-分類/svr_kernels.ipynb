# 房價預測

## 1. 載入相關套件

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 載入 Boston 房價資料集

```python
with open('./data/housing.data', encoding='utf8') as f:
    data = f.readlines()#讀入一行當作一個字串元素
all_fields = []
for line in data:#讀取一個字串(一行)
    line2 = line[1:].replace('   ', ' ').replace('  ', ' ')
    fields = []
    for item in line2.split(' '):
        fields.append(float(item.strip()))#字串去前後的空白-->轉小數-->加入到fields
        if len(fields) == 14:
            all_fields.append(fields)
df = pd.DataFrame(all_fields)
df.columns = 'CRIM,ZN,INDUS,CHAS,NOX,RM,AGE,DIS,RAD,TAX,PTRATIO,B,LSTAT,MEDV'.split(',')
df.head()
```

## 2. 資料清理、資料探索與分析

```python
# 觀察資料集彙總資訊
df.info()
```

```python
# 描述統計量
df.describe()
```

```python
# 是否有含遺失值(Missing value)
df.isnull().sum()
```

## 繪圖

```python
# 直方圖
import seaborn as sns

X, y = df.drop('MEDV', axis=1).values, df.MEDV.values
sns.histplot(x=y)
```

## 3. 不須進行特徵工程

## 4. 資料分割

```python
# 資料分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)

# 查看陣列維度
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 特徵縮放

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

## 5. 選擇演算法

```python
from sklearn.svm import SVR

model = SVR(kernel="linear") # "poly" ; "rbf"
```

## 6. 模型訓練

```python
model.fit(X_train_std, y_train)
```

## 7. 模型評分

```python
# R2、MSE、MAE
y_pred = model.predict(X_test_std)
print(f'R2 = {r2_score(y_test, y_pred)*100:.2f}') 
print(f'MSE = {mean_squared_error(y_test, y_pred)}') # mean squared error
print(f'MAE = {mean_absolute_error(y_test, y_pred)}') 
```

```python
# 取得偏差項及權重
model.intercept_, model.coef_
```
