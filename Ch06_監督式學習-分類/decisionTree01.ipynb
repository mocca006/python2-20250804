```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
wine = load_wine()#178,13
wine.data
```

```python
wine.data.shape
```

```python
wine.target
```

```python
# 合併欄位
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
# 分割資料
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data,
                                                wine.target,
                                                test_size=0.3)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

```python
Ytrain
```

```python
clf = tree.DecisionTreeClassifier(criterion="gini")#, max_depth=2
clf = clf.fit(Xtrain, Ytrain)

clf2 = tree.DecisionTreeClassifier(criterion="entropy")#, max_depth=2
clf2 = clf2.fit(Xtrain, Ytrain)

clf3 = tree.DecisionTreeClassifier(criterion="gini", max_depth=2)#葉子不算
clf3 = clf3.fit(Xtrain, Ytrain)

#返回預測的準確度accuracy
score = clf.score(Xtest, Ytest) 
score2 = clf2.score(Xtest, Ytest) 
score3 = clf3.score(Xtest, Ytest) 
score, score2, score3
```

```python
feature_name = ['酒精','蘋果酸','灰','灰的鹼性','鎂','總酚','類黃酮',
                '非黃烷類酚類','花青素','顏色強度','色調',
                'od280/od315稀釋葡萄酒','脯氨酸']

dot_data = tree.export_graphviz(clf, out_file="Tree01.dot"
                                #,feature_names = feature_name
                                ,feature_names = wine.feature_names
                                ,class_names=["0","1","2"]
                                ,filled=True
                                ,rounded=True
                               ) 

dot_data2 = tree.export_graphviz(clf2, out_file="Tree02.dot"
                                #,feature_names = feature_name
                                ,feature_names = wine.feature_names
                                ,class_names=["0","1","2"]
                                ,filled=True
                                ,rounded=True
                               ) 
dot_data3 = tree.export_graphviz(clf3, out_file="Tree03.dot"
                                #,feature_names = feature_name
                                ,feature_names = wine.feature_names
                                ,class_names=["0","1","2"]
                                ,filled=True
                                ,rounded=True
                               ) 
```

```python
# random_state隨機模式,多個維度時適用
clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=20)
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest) #返回評分結果(預測值)
score
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy"
                                    ,random_state=20
                                    ,splitter="random"
                                    ,max_depth=3 #層級
                                    ,min_samples_leaf=10  #葉子
                                    )
clf = clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
score
```
